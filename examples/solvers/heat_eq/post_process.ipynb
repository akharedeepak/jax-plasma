```python
import numpy as np
import xarray as xr
import seaborn, functools
import scipy
import matplotlib.pyplot as plt
```

```python
ds = xr.load_dataset("/home/deepak/Project/trial_short_project/jax-cfd/examples/solvers/heat_eq/my_dataset.nc")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Choose n evenly spaced time indices
n = 5  # or any number you want
time_indices = np.linspace(0, 
                           len(ds['time']) - 1, 
                           n, dtype=int)

plt.close()
# Create subplots
fig, ax = plt.subplots(1, n, figsize=(4*n, 4*1), constrained_layout=True)

vmin = -0.5#ds['T'][1:].min().item()
vmax = 0.5#ds['T'][1:].max().item()
for i, t_idx in enumerate(time_indices):
    im1 = ax[i].contourf(ds['x'], ds['y'], ds['T'][t_idx], 
                   levels=np.linspace(vmin, vmax, 20), 
                   cmap='viridis')
    ax[i].set_title(f"T at t={ds['time'].values[t_idx]:.2f}")

# Add shared colorbar
fig.colorbar(im1, ax=ax[i], orientation='vertical', label='Temperature')

plt.show()
```
